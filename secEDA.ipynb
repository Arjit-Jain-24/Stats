```python
df_1 = df_a.sample(frac=1).reset_index(drop=True)

print("Shuffled DataFrame:")


# Take samples and calculate means
num_samples = 300
sample_size = 300
sample_means1 = []


for _ in range(num_samples):
    sample = df_1.sample(n=sample_size, replace=False)  # Take a sample of size 30 without replacement
    sample_mean = sample['How much did the campaigning by the candidate\'s team(for whom you voted) influence your vote?'].mean()  # Calculate the mean of the specific column
    sample_means1.append(sample_mean)

# Calculate the mean and standard deviation of the means
mean_of_means1 = np.mean(sample_means1)
std_dev_of_means1 = np.std(sample_means1)

# Print the mean and standard deviation of the means
print("Mean of means:", mean_of_means1)
print("Standard deviation of means:", std_dev_of_means1)



print(sample_means1)
```

```python
import matplotlib.pyplot as plt

# Determine the range of sample means
min_value = min(sample_means1)
max_value = max(sample_means1)
range_value = (min_value, max_value)

# Plot histogram of sample means
plt.style.use('ggplot')  # Set the style of the plot
plt.hist(sample_means1, bins=10, range=range_value, edgecolor='black')  # Adjust the number of bins as needed
plt.xlabel('Sample Means')  # X-axis label
plt.ylabel('Frequency')     # Y-axis label
plt.title('Histogram of Sample Means')  # Title of the plot
# plt.grid(True)  # Show grid
plt.show()
```
