```python
# Group the data by 'Who did you vote for as President?' and calculate the mean for each group
means_by_candidate = df.groupby('Who did you vote for as President?')['How much did the campaigning by the candidate\'s team(for whom you voted) influence your vote?'].mean()

# Print the means for each candidate
print("Mean influence by candidate:")
print(means_by_candidate)

# Optionally, you can plot a bar chart to visualize the means
means_by_candidate.plot(kind='bar', xlabel='Candidate', ylabel='Mean influence', title='Mean Influence by Candidate')
```

```python
import pandas as pd

# Assuming your DataFrame is named 'df'
# Group the data by 'Who did you vote for as President?' and 'Why did you vote for him?'
grouped_data = df.groupby(['Who did you vote for as President?', 'Why did you vote for him?'])

# Count the number of people who chose a specific option for each candidate
counts_by_option = grouped_data.size()

# Calculate the total number of people who voted for each candidate
total_votes_by_candidate = df['Who did you vote for as President?'].value_counts().reset_index()
total_votes_by_candidate.rename(columns={'index': 'Who did you vote for as President?', 'Who did you vote for as President?': 'total_votes'}, inplace=True)
total_votes_by_candidate.set_index('Who did you vote for as President?', inplace=True)  # Set index to match counts_by_option

# Merge total_votes_by_candidate with counts_by_option to ensure the indices match
merged_data = pd.merge(counts_by_option, total_votes_by_candidate, left_index=True, right_index=True, how='outer')

# Calculate the percentage of people who chose a specific option for each candidate
merged_data['percentage'] = (merged_data[0] / merged_data['total_votes']) * 100

# Print the percentages for each candidate and option
print("Percentage of people by option and candidate:")
print(merged_data)

# Optionally, you can plot a bar chart to visualize the percentages
merged_data.reset_index(inplace=True)  # Reset index for plotting
merged_data.pivot(index='Why did you vote for him?', columns='Who did you vote for as President?', values='percentage').plot(kind='bar', stacked=True, xlabel='Reason for vote', ylabel='Percentage', title='Percentage by Option and Candidate')

```
