```python
# Two tailed_test
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

mean_before = np.mean(before_debate)
mean_after = np.mean(after_debate)
std_before = np.std(before_debate)
std_after = np.std(after_debate)
n = len(before_debate)
alpha = 0.01
# t_alpha = stats.t.ppf(1 - alpha/2, n-1)
# print(t_alpha)
# t = (mean_after - mean_before) / (std_after / np.sqrt(n))
# t_value = stats.t.cdf(t, n-1)
# print(t_value)
# p_value = stats.t.sf(np.abs(t), n-1) * 2
# print(p_value)
# Calculate test statistic
test_statistic = ((mean_after - mean_before) / std_after)* np.sqrt(n)

# Degrees of freedom
df = n - 1

# Find critical t-value for left-tailed test
critical_t_value = stats.t.ppf(alpha, df)

# Perform the hypothesis test
if test_statistic <= critical_t_value:
    print("Reject null hypothesis: std_after < std_before (left-tailed test)")
else:
    print("Fail to reject null hypothesis: std_after >= std_before (left-tailed test)")
```

```python
from scipy.stats import ttest_rel
```

```python
t_statistic, p_value = ttest_rel(after_debate, before_debate)

# Perform the hypothesis test
alpha = 0.01
if p_value < alpha:
    print("Rejected null hypothesis: std_after < std_before (left-tailed test)")
else:
    print("Fail to reject null hypothesis: std_after >= std_before (left-tailed test)")
```
