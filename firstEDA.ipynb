```python
import numpy as np
import pandas as pd
from scipy.stats import zipf, poisson
from scipy.optimize import minimize

# Step 1: Identify the Data


# Step 2: Choose a Zero-Inflated Distribution
# We'll use the Zero-Inflated Poisson distribution

# Step 3: Model the Zero-Inflation Component
# We'll use the Bernoulli distribution for the zero-inflation component

# Step 4: Model the Non-Zero Component
# We'll use the Poisson distribution for the non-zero component

# Step 5: Combine the Components
def zip_poisson_pmf(k, theta, mu):
    """
    Zero-Inflated Poisson PMF
    theta: Probability of zero-inflation
    mu: Mean of the Poisson distribution
    """
    mu = np.maximum(mu, 1e-8)  # Avoid zero values for mu
    theta = np.maximum(np.minimum(theta, 1 - 1e-8), 1e-8)  # Constrain theta between (0, 1)
    
    poisson_pmf = poisson.pmf(k, mu) * (1 - theta)
    zero_pmf = theta if k == 0 else 0
    return poisson_pmf + zero_pmf

# Step 6: Estimate Parameters
def zip_poisson_nll(params, data):
    """
    Negative Log-Likelihood function for Zero-Inflated Poisson distribution
    """
    theta, mu = params
    log_pmf = np.log([zip_poisson_pmf(x, theta, mu) for x in data])
    return -np.sum(log_pmf)

data = df['diffcol'].values
initial_params = [0.5, 1.0]  # Initial guesses for theta and mu
result = minimize(zip_poisson_nll, initial_params, args=(data,), method='L-BFGS-B')
theta_hat, mu_hat = result.x

# Step 7: Evaluate the Model
print(f"Estimated parameters: theta = {theta_hat:.4f}, mu = {mu_hat:.4f}")

# You can also plot the observed data and the fitted distribution for visual inspection
import matplotlib.pyplot as plt

counts, bins = np.histogram(data, bins=range(-10, 11), density=True)
bin_centers = 0.5 * (bins[:-1] + bins[1:])

fitted_pmf = [zip_poisson_pmf(k, theta_hat, mu_hat) for k in range(-9, 10)]

plt.figure(figsize=(8, 6))
plt.bar(bin_centers, counts, align='center', width=0.8, alpha=0.5, label='Observed Data')
plt.plot(range(-9, 10), fitted_pmf, 'r-', lw=2, label='Zero-Inflated Poisson')
plt.legend()
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Zero-Inflated Poisson Distribution Fit')
plt.show()

```
